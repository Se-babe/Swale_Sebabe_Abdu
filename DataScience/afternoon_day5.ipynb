```python
import matplotlib.pyplot as plt
 
x = [1,2,3,4,5]
y = [2,3,5,7,11]
plt.plot(x, y, marker='o', color='b', linestyle='-')
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```
